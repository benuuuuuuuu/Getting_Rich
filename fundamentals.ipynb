```python
import warnings
warnings.filterwarnings('ignore')

# columns we want
column_title = ['capxy', 'chechy', 'cibegniy', 'cicurry', 'cidergly',
       'cimiiy', 'ciothery', 'cipeny', 'cisecgly', 'citotaly', 'ciy', 'cogsy',
       'cshfdy', 'cshpry', 'cstkey', 'dilady', 'dilavy', 'doy', 'dvpy', 'dvy',
       'epsfiy', 'epsfxy', 'epspiy', 'epspxy', 'exrey', 'fiaoy', 'fincfy',
       'ibadjy', 'ibcomy', 'ibmiiy', 'iby', 'ivacoy', 'ivncfy', 'miiy', 'niy',
       'nopiy', 'oancfy', 'oepsxy', 'oiadpy', 'opepsy', 'piy', 'saley', 'spiy',
       'txbcofy', 'txty', 'xidoy', 'xiy', 'xopry']

# df to store matched data
match_df = pd.DataFrame(columns = column_title)

# loop through the data
from tqdm import tqdm
for i in tqdm(range(len(data))):
    tic = data['tic'].iloc[i]
    date = data['datadate'].iloc[i]
    tmp = f1[f1['tic'] == tic] 
    tmp = tmp[date > tmp['datadate']]
    if len(tmp)-1 >= 0:
        tmp = tmp.iloc[len(tmp)-1][column_title]
    else:
        tmp = pd.Series(data= float("NaN"), index=column_title)
        tmp.name = i
    match_df = match_df.append(tmp)
```
