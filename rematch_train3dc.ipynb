```
import pandas as pd
# price gram w2v
result = pd.read_csv('/content/drive/MyDrive/Social Media Group /Data/Train_result/price+gram+w2v.csv')

price = pd.read_csv('/content/drive/MyDrive/Social Media Group /Data/WRDS/close&eps_171819.csv')
```

```
# datetime transform 
price['datadate'] = pd.to_datetime(price['datadate'],format='%Y%m%d')
result['datadate'] = pd.to_datetime(result['datadate'],format='%Y-%m-%d')
```

```
price.columns
```

```
# price dict, key = ticker
ticker = price['tic'].drop_duplicates().tolist()
```

```
from tqdm import tqdm
price_dict = dict()
for tic in tqdm(ticker):
  price_dict[tic] = price[price['tic'] == tic].reset_index().reset_index()
```

```
result = result.drop(columns='3_day_change')
```

```
import warnings
warnings.filterwarnings('ignore')

# rematch 3 day change
from datetime import datetime, timedelta 

change_list = list()
for i in tqdm(range(len(result))):
  today_price = result['prccd'].iloc[i]
  tmp = price_dict[result['tic'].iloc[i]]
  today_date = result['datadate'].iloc[i]
  index = tmp[tmp['datadate'] == today_date]['level_0'].iloc[0]
  future_price = tmp['prccd'].iloc[index+3]
  change = (future_price-today_price)/today_price
  change_list.append(change)
```

```
len(change_list)
```

```
# match
result['3_day_change'] = change_list
```

```
result.to_csv('/content/drive/MyDrive/Social Media Group /Data/Train_result/price+gram+w2v_improved.csv')
```
