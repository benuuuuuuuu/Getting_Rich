# Topic Model-30

```
tp = pd.read_csv('/content/drive/MyDrive/Social Media Group /Data/result_topic30.csv')
```

```
# datetime transform
tp['datadate'] = pd.to_datetime(tp['datadate'],format='%Y-%m-%d')
```

```
column_name = list()
for i in range(30):
  tmp = 'topic'+str(i)
  column_name.append(tmp)
```

```
# str to list
import ast
tmp = list()
for index in tqdm(range(len(tp))):
    tmp.append(ast.literal_eval(tp['topicDistribution'].iloc[index]))
tp['topicDistribution'] = tmp
```

```
tp[column_name] = pd.DataFrame(tp['topicDistribution'].tolist())
```

```
tp.columns
```

```
tp = tp.set_index(['tic', 'datadate'])
```

```
# matching
match_tp = data.set_index(['tic','datadate']).join(tp, rsuffix='_topic')
```

```
match_tp = match_tp.reset_index()
match_tp = match_tp[column_name]
len(match_tp)
```

```
match_tp.to_csv('/content/drive/MyDrive/Social Media Group /Data/Train_result/matched_topics.csv')
```

# Indices indicators

```
import datetime
```

```
# s&p 500
spy = pd.read_csv('/content/drive/MyDrive/Social Media Group /Data/spy.csv')
spy = spy[['datadate','prccd']]
# datetime transform
spy['datadate'] = pd.to_datetime(spy['datadate'],format='%Y%m%d')
# index column
spy['S&P500'] = spy['prccd']
spy = spy.reset_index()
# match s&p with data df
tmp = list()
for index in range(len(data)):
    date = data['datadate'].iloc[index]
    try:
        spy_index = spy[spy['datadate'] == date]['index'].iloc[0]-1
    except:
        spy_index = spy[spy['datadate'] == date+datetime.timedelta(days=-1)]['index'].iloc[0]-2
    tmp.append(spy['S&P500'].iloc[spy_index])
# combine matches
data['sp500'] = tmp
```

```
# Russell 2000
rus = pd.read_csv('/content/drive/MyDrive/Social Media Group /Data/iwm.csv')
rus = rus[['datadate','prccd']]
# datetime transform
rus['datadate'] = pd.to_datetime(rus['datadate'],format='%Y%m%d')
# index column
rus['rus2000'] = rus['prccd']
rus = rus.reset_index()
# match s&p with data df
tmp = list()
for index in range(len(data)):
    date = data['datadate'].iloc[index]
    try:
        rus_index = rus[rus['datadate'] == date]['index'].iloc[0]-1
    except:
        rus_index = rus[rus['datadate'] == date+datetime.timedelta(days=-1)]['index'].iloc[0]-2
    tmp.append(rus['rus2000'].iloc[rus_index])
# combine matches
data['rus2000'] = tmp
```

```
# VIX
vix = pd.read_csv('/content/drive/MyDrive/Social Media Group /Data/vix.csv')
vix['Date'] = vix['Date'].apply(lambda x: x.upper())
vix['Date'] = pd.to_datetime(vix['Date'],format="%d%b%Y", yearfirst = False)
vix = vix.reset_index()
# match s&p with data df
tmp = list()
for index in range(len(data)):
    date = data['datadate'].iloc[index]
    try:
        vix_index = vix[vix['Date'] == date]['index'].iloc[0]-1
    except:
        vix_index = vix[vix['Date'] == date+datetime.timedelta(days=-1)]['index'].iloc[0]-2
    tmp.append(vix['vix'].iloc[vix_index])
# combine matches
data['vix'] = tmp
```
